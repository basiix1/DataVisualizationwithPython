#Data Visualization with Python

```python
import pandas as pd
```

```python
df = pd.read_csv('Canada.csv')
```

```python
df.head()
```

```python
df.drop(['Type','Coverage','AREA','REG','RegName','DEV','DevName'], axis=1 , inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={'OdName':'Country', 'AreaName':'Continent'},inplace=True)
```

```python
df.head()
```

```python
df.set_index('Country' ,inplace=True)
```

```python
df.head()
```

```python
years=list(map(str, range(1980, 2014)))
df.head()
```

```python
df['Total'] = df[df.columns[1:]].sum(axis=1)
df.head()
```

#### Exercise 2 : Line plot the top 5 countries that contributed the most to immigration to Canada

## Task 6
### Pie Chart

```python
df.head()
```

Group countries by continents

```python
Continents=df.groupby('Continent', axis=0).sum()
Continents
```

Create a pie chart to illustrate numerical proportion of immigrants from each continent

```python
colors_list=['brown','indianred','lightcoral','salmon','mistyrose','red']
explode_list= [0.2,0,0,0,0.2,0.2]
Continents['1988'].plot(kind='pie',
                       figsize=(20,10),
                       autopct='%1.1f%%',
                       startangle=90,
                       shadow=True,
                       labels=None,
                       pctdistance=1.12,
                       colors=colors_list,
                       explode=explode_list) 
plt.title("Immagration to Canada by Continent [1988]" , y=1.12)
plt.axis('equal')
plt.legend(labels=Continents.index ,loc='upper left')
plt.show()
```

## Task 8
### Sactter Plot

The correlation between two variables; years, and total number of immigrants

```python
df.head()
```

```python
df_Y=pd.DataFrame(df[years].sum(axis=0))
df_Y.reset_index(inplace=True)
df_Y.columns=['Year' ,'total']
df_Y.head()
```

```python
df_Y.plot(kind='scatter' , x='Year' , y='total' ,
            figsize=(10,7),
            color='green')

plt.title("Immagration -> Canada 1988 - 2013")
plt.xticks(rotation=90)
plt.ylabel("Total")
plt.xlabel("Y ears")
plt.show()
```
